### Задача 5
Брошены две игральные кости. Найти вероятность события D = {выпала хотя бы одна шестёрка}

Решение


А - появление шестёрки на первом кубике Р( А ) =1/6;
В - появление шестёрки на втором кубике Р (В ) =1/6.
События А и В совместны и независимы, применяем формулу
Р (А + В ) =Р (А ) + Р (В) - Р (А) Р( В ) =1/6 + 1/6 - 1/6. 1/6 =
11/36
Ответ 11/36

```python
k=100000

A = np.random.choice(range(1,7), k)
print(f'"A="{A}')
B = np.random.choice(range(1,7), k)
print(f'"B="{B}')

print(f'"Сумма AB"{A.tolist() + B.tolist()}')
print(f'"Произведение AB"{[a for a in A if a in B] + [b for b in B if b in A]}')
```

```python
#вероятнисти выигрыша и проигрыша
P_bingo = len(bingo)/len(choice)
P_fall = len(fall)/len(choice)
bingo_price = 10-1 #чистый выигрыш билета с номером 1
fall_price = 0-1 ##чистый выигрыш остальных билетов
```

```python
print("Таблица распределения случайной величины")

print("x =" , fall_price,"P(x) = " , P_fall)

print("x =" , bingo_price,"P(x) = " , P_bingo)
```

```python
print("Вероятность Х <= 4 =" , rel_freq[0]+rel_freq[1]+rel_freq[2] )
print("Вероятность Х > 4 =" , 1-(rel_freq[0]+rel_freq[1]+rel_freq[2]))
```

### Задача 10
Из урны, содержащей два белых и три черных шара, последовательно извлекают два шара; вынутые шары не возвращаются. Если первый шар оказался черным, то перед вторым извлечением в урну добавляют белый шар. В противном случае состав оставшихся шаров не меняют. Найти вероятности событий:

$A$ = {извлечены два белых шара},

$B$ = {извлечены два черных шара}

Указание. Пусть событие $C_1$ = {первый шар белый}, событие $C_2$ = {второй шар белый}. Тогда $A = C_1C_2$, $B = \overline C_1 \overline C_2$ 

```python
def fill_urn (n_white, n_black):
    return ["white"]*n_white + ['black']*n_black
```

```python
B1 = fill_urn(2,3)
B1
```

```python
A = {'white':0,'black':0}
```

```python
n= 100000
for i in range(n):
    #подходим к урне
    ball = random.choice(B1)
    #запоминаем из какой урны белый шар
    if ball == 'black':
        B1.append("white") 
    else:
        random.choice(B1)
    A.update({ball:A[ball]+1})  
```
