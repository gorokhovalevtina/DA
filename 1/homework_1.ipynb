```python
import csv
import sys
from collections import Counter, defaultdict
```

```python
# Открываем файл - в результате мы получаем следующую структуру данных - список со словарями,
#каждый элемент списка - это строка таблицы(словарь), ключи словаря - это названия колонок,
#а значения словаря - это значения в соответствующих ячейках таблицы
file='vacancy.csv'
vacs=list(csv.DictReader(open(file,encoding='utf-8')))
```

```python
#Посмотрим список вакансий
list=[x.get('vactitle') for x in vacs]
list
```

```python
# 1. Посчитаем количество вакансий, которые мне нравятся(Data - аналитик)
cnt = Counter(list)
cnt['Data - аналитик']
```

```python
# 2.1 Найдем, за какие даты эти вакансии
[x.get('vacdate') for x in vacs if x.get('vactitle').find('Data - аналитик')>-1]
```

```python
# 2.2 Посмотрим описание этой вакансии
[x for x in vacs if x.get("vactitle").find("Data - аналитик")>-1]
```

```python
# 3. Посчитаем, сколько вакансий с такими позициями, на которой работаю я (Руководитель дополнительного офиса)
cnt['Руководитель дополнительного офиса']
```

```python
# 4.1 Посчитаем, сколько вакансий для аналитика данных (Строго в таком написании вакансии)
cnt['Аналитик данных']
```

```python
# 4.2 количество вакансий, где есть слова "аналитик данных"
len([x for x in vacs if x.get("vactitle").lower().find("аналитик данных")>-1])
```

```python
# 4.3 описание этих вакансий
[x for x in vacs if x.get("vactitle").lower().find("аналитик данных")>-1]
```

```python
# 4.4 даты этих вакансий
[x.get('vacdate') for x in vacs if x.get('vactitle').lower().find('аналитик данных')>-1]
```

```python
#, сколько вакансий для аналитика данных с использованием Python
len([x for x in vacs if x.get("vacdescription").lower().
     find("python")>-1 and x.get('vactitle').lower().find('аналитик данных')>-1])
```
